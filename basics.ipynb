**Confusion Matrix**

Table used to evaluate performace of a classification model by comparing actual labels with predicted labels


```python
from sklearn.metrics import confusion_matrix
y_true = [1, 0, 1, 1, 0, 1, 0, 0]
y_pred = [1, 0, 1, 0, 0, 1, 1, 0]

cm = confusion_matrix(y_true, y_pred)
print(cm)
```
